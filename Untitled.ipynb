```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
data=pd.read_csv('hccdata.csv')
```

```python
temp={"alcohol":data['Alcohol'],"ciirhosis":data['Cirrhosis'],"smoking":data['Smoking'],"obesity":data['Obesity']}
```

```python
numberOfAlcoholi=np.sum(temp['alcohol'])
numberOfCiirhosis=np.sum(temp['ciirhosis'])
numberOfSmoking=np.sum(temp['smoking'])
numberOfObesity=np.sum(temp['obesity'])
```

```python
print "Numner of Alcoholic people "+str(numberOfAlcoholi)
print "Numner of Ciirhosis people "+str(numberOfCiirhosis)
print "Numner of Smoking people "+str(numberOfSmoking)
print "Numner of Obesed people "+str(numberOfObesity)
```

```python
#ferquency table
labels=['alcohol','ciirhosis','smoking','obesity']
freq=[numberOfObesity,numberOfAlcoholi,numberOfCiirhosis,numberOfAlcoholi+numberOfCiirhosis,numberOfAlcoholi+numberOfSmoking,numberOfAlcoholi+numberOfObesity,numberOfSmoking]
x=[1,2,3,4,5,6,7]
freq=np.array(freq,dtype=float)
```

```python
print(freq)
```

```python
probability=freq/np.sum(freq)
```

```python
print("X  Probability")
print("-----------------")
for i in range(freq.shape[0]):
    print(str(x[i])+" | "+str(probability[i]))
```

```python
plt.plot(x,freq)
plt.xlabel("Value of X")
plt.ylabel("Frequency")
plt.show()
```

```python
#mean median mode
mean=0.0
meadian=0.0
mode=0.0

for i in range(freq.shape[0]):
    mean=mean+x[i]*probability[i]
cf=[]
t=0
for i in range(freq.shape[0]):
    t=t+freq[i]
    cf.append(t)
n=cf[6]

for i in range(freq.shape[0]-1):
    if(n/2>cf[i] and n/2<cf[i+1]):
        median=i
median=x[median]

t=freq[0]

for i in range(1,freq.shape[0]):
    if(t<freq[i]):
        t=freq[i]
        mode=i
mode=x[mode]


```

```python
plt.plot(x,probability)
plt.ylabel("Probability density function")
plt.xlabel("Value of x")
plt.show()

temp=[]
for i in range(freq.shape[0]):
    for j in range(100):
        temp.append([x[i]+j*0.01,cf[i]/np.sum(freq)])
    
temp=np.array(temp)
plt.scatter(temp.T[0],temp.T[1])
plt.ylabel("Distribution function")
plt.xlabel("Value of x")
plt.show()

```

```python
print("Mean "+str(mean))
print("Median "+str(median))
print("Mode "+str(mode))
```

```python
#variance
var=0.0
sd=0.0
for i in range(freq.shape[0]):
    var=var+x[i]*x[i]*probability[i]
var=var-mean*mean
sd=math.sqrt(var)
```

```python
print("Variance "+str(var))
print("Standard deviation "+str(sd))
```
